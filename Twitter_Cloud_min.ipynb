```python
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
# open the file and read it into a variable twitter_desc
twitter_desc = open('/Users/marco/Documents/Private/Coursera/Data_Visualization/twitter_desc.txt', 'r').read()
print(twitter_desc)
print ('File downloaded and saved!')
```

```python
stopwords = set(STOPWORDS)
```

```python
stopwords.add('Twitter') # add the words said to stopwords
```

```python
# instantiate a word cloud object
twitter_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

# generate the word cloud
twitter_wc.generate(twitter_desc)
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

# display the cloud
plt.imshow(twitter_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('tweets') # add the words said to stopwords

# re-generate the word cloud
twitter_wc.generate(twitter_desc)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(twitter_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
